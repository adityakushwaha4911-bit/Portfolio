```
from sklearn.impute import SimpleImputer
```

```
data6 = pd.read_csv('/content/drive/MyDrive/LAB2_Preprocessing_Dataset.csv')
```

```
print("original Dataset:")
print(data6)
```

```
numerical_data = data6[['Age','Salary']]
```

```
print("\nNumerical Data with Missing Values:")
print(numerical_data)
```

```
imputer = SimpleImputer(strategy='mean')
```

```
processed_numerical_data = imputer.fit_transform(numerical_data)
```

```
processed_numerical_df = pd.DataFrame(processed_numerical_data, columns=numerical_data.columns)
```

```
print("\nNumerical Data after Imputation:")
print(processed_numerical_df)
```
